```python
from biz_opps.neo4j.helpers import get_neo4j_driver
from biz_opps.etl.block_groups import get_block_group_data
from biz_opps.utils.geometry import convert_crs
```

```python
neo4j_driver = get_neo4j_driver()
block_group_gdf = get_block_group_data(suffix="LIMIT 1")
```

```python
# Determine the crs of the geometry
crs = block_group_gdf.crs
print(f"CRS of the GeoDataFrame: {crs}")
```

```python
# Convert the geometry to 4326
block_group_gdf = convert_crs(block_group_gdf)
```

```python
# Select the first geometry
shape = block_group_gdf.wkb_geometry.iloc[0]

# plot the geometry on a world map
import folium
import shapely.wkb

# plot the geometry on a world map
m = folium.Map(location=[shape.centroid.y, shape.centroid.x], zoom_start=12)
folium.GeoJson(shape).add_to(m)
m
```
